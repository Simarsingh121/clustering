 **Question 1**: Difference between K-Means and Hierarchical Clustering

Answer:

- K-Means and Hierarchical Clustering are both clustering techniques, but they work in different ways.

- K-Means clustering divides the data into a fixed number of clusters (K). It works by assigning each data point to the nearest cluster center and then updating the cluster centers repeatedly until they stabilize. It is fast and works well when the number of clusters is already known.

- Hierarchical clustering does not require us to decide the number of clusters in advance. It builds a hierarchy of clusters either by merging smaller clusters (agglomerative) or splitting larger clusters (divisive). The result is shown using a dendrogram, which helps decide the number of clusters later.

Use case:

- K-Means is used in customer segmentation when we already know how many customer groups we want.

- Hierarchical clustering is useful in biology (gene classification) where the relationship between data points matters.

**Question 2**  Purpose of Silhouette Score

Answer:

- The Silhouette Score is used to measure how well data points are clustered. It tells us how similar a data point is to its own cluster compared to other clusters.

- The score ranges from -1 to +1:

- A value close to +1 means the data point is well clustered.

- A value near 0 means the point lies between clusters.

- A negative value means the point may be in the wrong cluster.

- It helps in comparing different clustering results and choosing the best number of clusters.

**Question 3**: Core parameters of DBSCAN

Answer:

- DBSCAN mainly depends on two parameters:

eps (epsilon):
- It defines the maximum distance between two points to be considered neighbors. A small eps may create too many small clusters, while a large eps may merge clusters.

min_samples:
- It is the minimum number of points required to form a dense region. Higher values make clustering stricter.

- Together, these parameters decide how clusters and noise points are identified.

**Question 4:** Importance of Feature Scaling

Answer:

- Feature scaling is important because clustering algorithms like K-Means and DBSCAN use distance calculations.

- If features are on different scales (for example, age vs income), larger values dominate the distance calculation and give wrong clusters.

- Scaling ensures all features contribute equally, leading to better and more meaningful clusters.

**Question 5**: Elbow Method in K-Means

Answer:

 - The Elbow Method helps determine the optimal number of clusters in K-Means.

- It works by plotting the number of clusters (K) against the Within-Cluster Sum of Squares (WCSS).
- As K increases, WCSS decreases. The point where the decrease slows down forms an “elbow”.

That elbow point is considered the best value for K.

**Question 6:** Generate synthetic data using make_blobs(n_samples=300, centers=4),
apply KMeans clustering, and visualize the results with cluster centers.

**Output Explanation:**
The plot shows 4 clear clusters with red X marks representing the cluster centers.


```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=300, centers=4, random_state=42)

kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1],
            color='red', marker='x')
plt.show()
```

**Question 7:** Load the Wine dataset, apply StandardScaler , and then train a DBSCAN
model. Print the number of clusters found (excluding noise).


```python
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import numpy as np

wine = load_wine()
X = wine.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

dbscan = DBSCAN(eps=1.5, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print("Number of clusters:", n_clusters)
```

**Question 9:** Load the Wine dataset, reduce it to 2D using PCA, then apply
Agglomerative Clustering and visualize the result in 2D with a scatter plot.

**Output Explanation:**
The wine data is reduced to 2D and grouped into 3 clusters for easy visualization.



```python
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

wine = load_wine()
X = wine.data

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

agg = AgglomerativeClustering(n_clusters=3)
labels = agg.fit_predict(X_pca)

plt.scatter(X_pca[:,0], X_pca[:,1], c=labels)
plt.show()
```
